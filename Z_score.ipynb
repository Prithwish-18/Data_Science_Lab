```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt
```

```python
passport_df=pd.read_csv('passport.csv')
passport_df.head(5)
```

```python
plt.hist(passport_df["Processing time"], bins=10)
plt.xlabel("Processing Time (Days)")
plt.ylabel("Frequency")
plt.title("Histogram of Passport Processing Time")
plt.show()

mean = passport_df["Processing time"].mean()
std = passport_df["Processing time"].std()

print("Sample Mean:", mean)
print("Sample Standard Deviation:", std)

import math
from scipy import stats

pop_mean = 30
pop_std = 12.5
n = len(passport_df)

z_score = (mean - pop_mean) / (pop_std / math.sqrt(n))
p_value = stats.norm.cdf(z_score)

print("Z-score:", z_score)
print("P-value:", p_value)
```
